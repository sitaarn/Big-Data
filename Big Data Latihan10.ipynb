```python
# Latihan 1

from pymongo import MongoClient

# Inisialisasi client dan koneksi ke database
client = MongoClient('mongodb://localhost:27017/')
db = client['company_db']
collection = db['employees']

# Contoh untuk memastikan koneksi
print('Connected to MongoDB')
```

```python
# Latihan 2

# Insert satu data
employee_data = {
    'name': 'Alice',
    'department': 'Finance',
    'age': 29,
    'salary': 4500
}
collection.insert_one(employee_data)
print("1 data berhasil dimasukkan")
```

```python
# Tugas 2

import pandas as pd
import random

# Generate 120 data karyawan acak 
names = ["Bob", "Charlie", "Diana", "Eve", "Frank", "Grace", "Henry", "Ivy", "Jack"]
departments = ["Finance", "IT", "HR", "Marketing", "Sales"]
data = []

for i in range(120):
    employee = {
        "name": random.choice(names),
        "department": random.choice(departments),
        "age": random.randint(22, 50),
        "salary": round(random.uniform(3500, 9000), 2)
    }
    data.append(employee)

# Insert ke MongoDB 
collection.insert_many(data)
print(f"{len(data)} data karyawan berhasil dimasukkan\n")

# Update dan Delete contoh 
# Tambah 5% gaji untuk semua karyawan IT
collection.update_many({'department': 'IT'}, {'$mul': {'salary': 1.05}})

# Hapus semua karyawan dengan gaji < 4000
collection.delete_many({'salary': {'$lt': 4000}})

# Ambil semua data hasil akhir 
final_data = list(collection.find({}, {"_id": 0}))

# Ubah jadi DataFrame 
df = pd.DataFrame(final_data)

# Tampilkan hasil tabel 
print("Data Karyawan Setelah Operasi Update & Delete:\n")
display(df)

# Tampilkan ringkasan jumlah per departemen 
print("\nJumlah karyawan per departemen:")
print(df["department"].value_counts())
```

```python
# Tampilan dalam bentuk tabel

import pandas as pd

# Ambil semua data dari koleksi tanpa _id
data = list(collection.find({}, {'_id': 0}))

# Ubah jadi DataFrame biar tampil seperti tabel
df = pd.DataFrame(data)

# Tampilkan hasil tabel
print("Data Karyawan:")
display(df)
```

```python
# Latihan 3 (Dalam bentuk JSON)

# Query Aggregation untuk mencari rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
# Dalam bentuk tabel 

import pandas as pd

# Aggregation: hitung rata-rata gaji per departemen
pipeline = [
    {'$group': {'_id': '$department', 'average_salary': {'$avg': '$salary'}}}
]
results = list(collection.aggregate(pipeline))

# Ubah hasil ke DataFrame
df = pd.DataFrame(results)

# Ubah nama kolom agar lebih jelas
df.rename(columns={'_id': 'Department', 'average_salary': 'Average Salary'}, inplace=True)

# Tampilkan tabel hasil rata-rata gaji
print("üìä Rata-rata Gaji per Departemen:")
display(df)
```

```python
# Latihan 4.1 (Buatlah koleksi baru products dan masukkan data produk (minimal 10 produk).

from pymongo import MongoClient
import pandas as pd

# Koneksi ke MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["company_db"]
products = db["products"]

# Data produk minimal 10 item
product_list = [
    {"name": "Laptop ASUS", "category": "Electronics", "price": 12000000, "stock": 10},
    {"name": "Smartphone Samsung", "category": "Electronics", "price": 8000000, "stock": 20},
    {"name": "Headphone Sony", "category": "Accessories", "price": 1500000, "stock": 50},
    {"name": "Keyboard Logitech", "category": "Accessories", "price": 750000, "stock": 40},
    {"name": "Mouse Razer", "category": "Accessories", "price": 650000, "stock": 35},
    {"name": "Kursi Gaming Secretlab", "category": "Furniture", "price": 6000000, "stock": 5},
    {"name": "Meja Kantor IKEA", "category": "Furniture", "price": 3500000, "stock": 8},
    {"name": "Monitor LG", "category": "Electronics", "price": 2500000, "stock": 15},
    {"name": "Printer Canon", "category": "Electronics", "price": 2200000, "stock": 12},
    {"name": "Webcam Logitech", "category": "Accessories", "price": 950000, "stock": 30}
]

# Insert ke koleksi
products.insert_many(product_list)
print("10 data produk berhasil dimasukkan!")

# Tampilkan semua data dalam tabel
df_products = pd.DataFrame(list(products.find({}, {'_id': 0})))
display(df_products)
```

```python
# Latihan 4.2 (Lakukan query untuk menemukan produk dengan harga di atas rata-rata.)

# Hitung rata-rata harga
avg_price = list(products.aggregate([
    {'$group': {'_id': None, 'average_price': {'$avg': '$price'}}}
]))[0]['average_price']

print(f"Rata-rata harga produk: Rp{avg_price:,.0f}\n")

# Cari produk dengan harga di atas rata-rata
above_avg = list(products.find({'price': {'$gt': avg_price}}, {'_id': 0}))

# Tampilkan hasil dalam tabel
df_above_avg = pd.DataFrame(above_avg)
print("Produk dengan harga di atas rata-rata:")
display(df_above_avg)
```

```python
# Latihan 4.3 (Buatlah aggregation pipeline untuk menghitung total produk dalam setiap kategori.)

# Pipeline aggregation
pipeline = [
    {'$group': {'_id': '$category', 'total_products': {'$sum': 1}}}
]

category_counts = list(products.aggregate(pipeline))

# Tampilkan hasil dalam bentuk tabel
df_category = pd.DataFrame(category_counts)
df_category.rename(columns={'_id': 'Category', 'total_products': 'Total Products'}, inplace=True)

print("Jumlah produk per kategori:")
display(df_category)
```
